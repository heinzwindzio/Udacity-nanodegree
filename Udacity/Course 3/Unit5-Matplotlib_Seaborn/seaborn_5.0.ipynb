```python
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# Matplotlib
plt.scatter(x=[1, 2, 3], y=[4, 5, 6])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Matplotlib Scatter')
plt.show()

# Seaborn
sb.scatterplot(x=[1, 2, 3], y=[4, 5, 6])
plt.title('Seaborn Scatter')
plt.show()
```
