```python
# Necessary imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# Read the csv file as a panda DataFrame, and check its top 10 rows
pokemon = pd.read_csv('pokemon.csv')
print(type(pokemon))
print(pokemon.shape)
pokemon.head(10)

```

### to make a bar chart, provide seaborn with the panda DataFrame and the x axis to create it ###

then call matplotlib to show the bar chart


```python

sb.countplot(data=pokemon, x='generation_id')
plt.show()


```

### to change the color, grab the color from seaborn and then provide it ###

```python
base_color = sb.color_palette()[1]

sb.countplot(data=pokemon, x='generation_id', color=base_color)
plt.show()
```

### use matplotlib instead of seaborn to create a bar chart ###

```python
# Return unique values for x
x = pokemon['generation_id'].unique()

# Return the frequency counts for each unique value
y = pokemon['generation_id'].value_counts(sort=False)

# create the bar chart with the x and y axis - no need to provide a DataFrame
plt.bar(x, y)

# Labeling the axes
plt.xlabel('generation_id')
plt.ylabel('count')

# Dsiplay the plot
plt.show()
```

```python
# Order the bars
sb.countplot(data=pokemon, x='generation_id', color=base_color, order=[5,1,3,4,2,7,6]);
plt.show()
```

### dynamic ordering of the bars ###

```python
# grab the panda series representing the counts for each generation_id
freq = pokemon['generation_id'].value_counts()
print(type(freq))
print(freq)

# Get the indexes of the Series - 
gen_order = freq.index

# Plot the bar chart in the decreasing order of the frequency of the `generation_id`
sb.countplot(data=pokemon, x='generation_id', color=base_color, order=gen_order);

plt.show()
```

```python
# Plot the Pokemon type on a Vertical bar chart
sb.countplot(data=pokemon, x='generation_id', color=base_color);

# Use xticks to rotate the category labels (not axes) counter-clockwise
plt.xticks(rotation=90)
plt.show()
```
